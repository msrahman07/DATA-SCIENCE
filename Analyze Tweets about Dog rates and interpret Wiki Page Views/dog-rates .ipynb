## Pup Inflation: Analysing Tweets

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from datetime import datetime
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
dog_rates_df = pd.read_csv('dog_rates_tweets.csv', parse_dates=['created_at'], date_parser=dateparse)
```

```python
import re
rates_re = re.compile(r'(\d+(\.\d+)?)/10')
def get_rates(txt):
    match = rates_re.search(txt)
    if match:
        return match.group(1)
    else:
        return None

dog_rates = dog_rates_df['text'].apply(get_rates).dropna()
dog_rates_float = dog_rates.apply(float)
dog_rates_df['rating'] = dog_rates_float
dog_rates_df = dog_rates_df[dog_rates_df['rating'] <= 25].dropna()
```

```python
dog_rates_df = dog_rates_df.dropna()
```

## Linear Fitting

```python
dog_rates_df['timestamp'] = dog_rates_df['created_at'].apply(datetime.timestamp)
```

```python
dog_rates_df
```

```python
fit = stats.linregress(dog_rates_df['timestamp'].values, dog_rates_df['rating'].values)
```

```python
fit.slope, fit.intercept
```

```python
plt.xticks(rotation=25)
plt.title('Pup Inflation & Linear Fitting')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.plot(dog_rates_df['created_at'].values, dog_rates_df['rating'].values,'b.', alpha=0.5)
plt.plot(dog_rates_df['created_at'].values, (dog_rates_df['timestamp'].values*fit.slope+fit.intercept), 'r-', linewidth = 3)
plt.savefig('dog-rates-result.png')
```
