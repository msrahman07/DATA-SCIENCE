```python
import pandas as pd
```

```python
data = pd.read_csv('precipitation.csv', parse_dates=[2])
```

```python
data
```

```python
def date_to_month(d):
    # You may need to modify this function, depending on your data types.
    return '%04i-%02i' % (d.year, d.month)
```

```python
data['month'] = data['date'].apply(date_to_month)
```

```python
grouped_data = data.groupby(['name', 'month'])
```

```python
monthly_new_data = grouped_data.aggregate('sum').reset_index()
```

```python
monthly_new_data
```

```python
monthly = monthly_new_data.pivot(index='name', columns='month', values='precipitation')
```

```python
monthly
```

```python
counts_new_data = grouped_data.aggregate('count').reset_index()
```

```python
counts = counts_new_data.pivot(index='name', columns='month', values='precipitation')
```

```python
counts
```
